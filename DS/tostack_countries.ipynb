```python
 import pandas as pd
```

```python
#delete unnecessary columns

df = pd.read_csv('new_bycountry.csv', index_col="country_name")
del df["Unnamed: 0"]
del df["continents"]
del df["population_number_of_people"]
```

```python
#look for the difference to reach 100 
df['difference'] = round(100 - df.sum(axis=1),1)
```

```python
df["other_waste_percent"] = df["other_waste_percent"] + df["difference"]
```

```python
del df['difference']
```

```python
#renaming columns for the visualization 

cols2 = { 'organic_food_percent': 'Organic',	'glass_percent': 'Glass',	'metal_percent': 'Metal',	'other_waste_percent': 'Other',	'paper_cardboard_percent': 'Paper',	'plastic_percent': 'Plastic'}

df = df.rename(columns=cols2)
```

```python
df
```

```python
#unstacking the data frame so we can use it for the bar graph

df_test = df.unstack(0).reset_index()
```

```python
#creating a list to rename columns

cols = {"level_0":"type_of_waste", 0:"percentage"}
```

```python
#renaming columns

df_test = df_test.rename(columns=cols)
```

```python
df_test.to_csv("unstacked_df_v1.csv")
```
