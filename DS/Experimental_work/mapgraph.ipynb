```python
import pandas as pd

#use the final data set with predictions and real data 

wastebycountry= pd.read_csv('mapgraph.csv')
```

```python
wastebycountry
```

```python
#create a new data frame with only the "countr_name" and "2020" columns
data2020= wastebycountry[["country_name", "2020"]]
```

```python
#change the name to Germany to match with the map
data2020.at[13,'country_name']='Germany'
```

```python
#change the name to Kosovo to match with the map
data2020.at[19,'country_name']='Kosovo'
```

```python
data2020.to_csv('data2020.csv')
```

```python
data2020
```

```python
#set the index of the data dataframe to the country name
data2020=data2020.set_index('country_name')
```

```python
#drop the row European union-...
data2020=data2020.drop('European Union - 27 countries (from 2020)')
```

```python
data2020
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

#read in the shapefile for Europe
eu_map = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
eu_map = eu_map[eu_map.continent == 'Europe']
```

```python
import matplotlib as mpl

#merge the data with the shapefile
eu_data = eu_map.merge(data2020, left_on='name', right_index=True)

#create a choropleth map
fig, ax = plt.subplots(figsize=(12, 8),dpi=200)
eu_data.plot(column='2020', cmap='YlOrRd', ax=ax, legend=True, edgecolor='white')
ax.set_title('Waste generated per person in kg')
ax.set_axis_off()
ax.set_xlim([-20, 40]) # set the longitude limits for the map
ax.set_ylim([30, 70])  # set the latitude lgimits for the map


plt.show()
```
