```python
#create a new column with country code in order to later create a column with the name of the continent
#import packages to create a function 

import pycountry as pc 

```

```python
import country_converter as coco
import pycountry_convert as pc1
```

```python
#create a function to create the continent column in the data frame


def country_to_continent(country_name):
    try:
        country_alpha2 = pc1.country_name_to_country_alpha2(country_name)
        country_continent_code = pc1.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc1.convert_continent_code_to_continent_name(country_continent_code)
    except:
        country_continent_name = ""
    return country_continent_name

```

```python
#Creating the new column applying the country_to_continent function

bycountry_produced_cleaned['continents']= bycountry_produced_cleaned['country_name'].apply(country_to_continent)
```
