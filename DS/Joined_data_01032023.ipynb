```python
#INNER JOIN between 2 DataFrames--> set of records that have a mutual value in the specified joining column(s)
joined=waste.merge(recycle, on =["country_name", "region_id", "year"])
joined.head(30)
```

```python
#replace zero value with NAN
joined.replace(0, np.nan, inplace=True)
joined.head(30)
```

```python
#delete missing values
joined.dropna(inplace=True)
joined.head(40)
```

```python
app = Dash(__name__)
```

```python
#create #dropdown list for Dataset 1
app.layout = html.Div([
    html.H1('Waste Data Visualization'),
    dcc.Dropdown(id='year_dropdown',
                 value=2002,
                 options=[{'label': year, 'value': year}
                          for year in list(joined["year"].unique())]),
    dcc.Graph(id='waste_chart'),
])
```

```python
@app.callback(Output('waste_chart', 'figure'),
              Input('year_dropdown', 'value'))
def plot_teams_total(year):
    
    year_joined = joined[joined["year"] == year].sort_values("municipal_waste_per_capita_kg", ascending=False)
    len_countries = len(year_joined["NOC"].unique())   
    
    fig = px.bar(
        year_joined,
        x='municipal_waste_per_capita_kgl',
        y='country_name',
       
        height=20 + (15 * len_countries),
        title= f'Countries & Waste per capita - {year}',
    )
    fig.update_layout(yaxis_categoryorder='total ascending')
    
    return fig
```
